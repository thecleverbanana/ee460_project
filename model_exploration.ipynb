```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from glob import glob
from IPython.display import Image
```

RBF on expanding window with fixed val/test data frame (2022-2023/2023-2024)

```python
metrics_dir = "rbf_ewf/expanding"
all_files = glob(os.path.join(metrics_dir, "*_metrics.csv"))
```

```python
df_list = []
for file in all_files:
    df = pd.read_csv(file)
    df["Train Start"] = pd.to_datetime(df["Train Start"]).dt.year
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)
df_all.sort_values("Train Start", inplace=True)

df_all.to_csv(os.path.join(metrics_dir, "all_metrics.csv"), index=False)
```

```python
plot_dir = os.path.join(metrics_dir, "plots_summary")
os.makedirs(plot_dir, exist_ok=True)

metrics_to_plot = [
    "Val R2", "Test R2",
    "Validation Sharpe Ratio", "Test Sharpe",
    "Validation Cumulative Return", "Test Return",
    "Final Val Capital", "Final Test Capital",
    "Total Val Profit", "Total Test Profit",
    "Average PnL", "Average PnL (%)"
]

for metric in metrics_to_plot:
    plt.figure(figsize=(10, 5))
    sns.lineplot(x="Train Start", y=metric, data=df_all, marker="o")
    plt.title(f"{metric} over Train Start")
    plt.xlabel("Train Start Year")
    plt.ylabel(metric)
    plt.grid(True)
    plt.tight_layout()
    save_path = os.path.join(plot_dir, f"{metric.replace(' ', '_')}.jpg")
    plt.savefig(save_path, dpi=300)
    plt.close()
    print(f"✓ Saved plot: {save_path}")
```
